```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn import preprocessing
from scipy import stats
```

```python
data = pd.read_csv("Electric.csv")
data = data["Electric"].tolist()
data = stats.zscore(np.array(data))
data = data.tolist()
```

```python
plt.plot(data)
plt.show()
```

```python
seq=[]
for  i in range(0, len(data), 50) :
    if (i+3072) > len(data):
        break
    array=np.asarray(data[i:i+3072])
    seq.append(array)
seq=np.asarray(seq)
```

```python
seq=seq.reshape(len(seq), 3072)
seq=seq[:3072,::]
```
